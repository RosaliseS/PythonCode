This project aims to get a clear understanding of how to use conditional statements to make a

# Story 
Rohan is bored wearing the jacket and pullover now. He wants to wear something light and soft. 
But he also doesn't want to get the cold by not wearing weather-appropriate clothes. So he will 
design a program and check what temperature is suitable for wearing light clothes.

```python
temperature = float(input("What is the temperature? "))
if temperature >= 18:
    print("It's OK to wear summer clothes!")
else: print("It's too cold to wear summer clothes") 
```
